***Problem 2:*** <br>
Scrape the descriptions of all 20 books that are on this page: http://books.toscrape.com/ .<br>The descriptions of books can be accessed by clicking on the book image or title and going to its personal page. <br>
**Note:** Please, use  BeautifulSoup.

```python
###_____________Start code here_________________###
books="http://books.toscrape.com/"
wd=webdriver.Chrome()
wd.get(books)
page=requests.get(books)
page.status_code
ej=page.content
ej=BeautifulSoup(ej,'html.parser')
h3=ej.find_all("h3")
hrefs=[i.findChild("a")["href"] for i in h3]
description=[]

for i in hrefs:
    response=requests.get('http://books.toscrape.com/'+i)
    page=BeautifulSoup(response.content,"html.parser")
    div=page.find("div",id="product_description",class_="sub-header")
    p=div.findNext("p")
    description.append(p.get_text())
    
    

print(description)


###_____________End code here_________________###
```

***Problem 3:*** <br>
Scrape  all "Tags" from the website http://quotes.toscrape.com/scroll.<br>
For example, scrape change, deep-thoughts, thinking world from the first quote.
**Note:** You need to scroll down untill the end of the page using Selenium. <br>

```python
###_____________Start code here_________________###

#scrolling down the page
url="http://quotes.toscrape.com/scroll"
wd=webdriver.Chrome()
wd.get(url)
last_height = wd.execute_script("return document.body.scrollHeight")
while True:
    wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    new_height = wd.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
  

pg=wd.page_source 
pg=BeautifulSoup(pg,'html.parser')
tags=pg.find_all("a", class_="tag")
tagnames=[]
for i in tags:
    tagnames.append(i.get_text())
    
set(tagnames)

print(tagnames)





###_____________End code here_________________###
```
